```python
import pandas as pd 
import numpy as np
```

```python
# Define parameters for fresh data generation
num_regions = 20
num_years = 3
num_weeks = num_years * 52
start_date = pd.to_datetime("2021-01-01")

# Generate fresh dummy marketing data
data_new = []
for region in range(1, num_regions + 1):
    for week in range(num_weeks):
        week_start = start_date + pd.DateOffset(weeks=week)
        sales = np.random.randint(500, 5000) 
        # HCP Calls
        hcp_calls = np.random.randint(50, 500)
        hcp_calls_spend = hcp_calls * np.random.randint(100, 200)  # increased per-call cost

        # HCP Samples
        hcp_samples = np.random.randint(20, 200)
        hcp_samples_spend = hcp_samples * np.random.randint(150, 250)  # per-sample cost is generally higher due to material/shipping

        # HCP Print
        hcp_print = np.random.randint(5, 50)
        hcp_print_spend = hcp_print * np.random.randint(500, 1000)  # printing costs tend to be bulk per order, so higher per unit

        # DTC TV
        dtc_tv = np.random.randint(1000, 10000)
        dtc_tv_spend = dtc_tv * np.random.randint(200, 400)  # higher cost per GRP/impression for TV

        # DTC Display
        dtc_display = np.random.randint(500, 7000)
        dtc_display_spend = dtc_display * np.random.randint(20, 50)  # relatively lower cost per impression

        
        data_new.append([f"Region_{region}", week_start, sales, hcp_calls, hcp_calls_spend, hcp_samples, hcp_samples_spend, 
                         hcp_print, hcp_print_spend, dtc_tv, dtc_tv_spend, dtc_display, dtc_display_spend])

# Create new DataFrame
columns_new = ["Geo ID", "Week start date", "Sales", "HCP Calls", "HCP Calls Spend", "HCP Samples", "HCP Samples Spend", 
               "HCP Print", "HCP Print Spend", "DTC TV", "DTC TV Spend", "DTC Display", "DTC Display Spend"]
df_new = pd.DataFrame(data_new, columns=columns_new)

# Add 'Month' and 'Year' columns
df_new['Month'] = df_new['Week start date'].dt.month
df_new['Year'] = df_new['Week start date'].dt.year

# Reorder columns to place Month and Year next to Week start date
cols = df_new.columns.tolist()
date_index = cols.index("Week start date")
# Insert Month and Year after Week start date
cols.insert(date_index + 1, cols.pop(cols.index("Month")))
cols.insert(date_index + 2, cols.pop(cols.index("Year")))
df_new = df_new[cols]

# Display sample data
df_new.head()
```
